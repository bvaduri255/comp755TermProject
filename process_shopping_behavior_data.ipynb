```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("shopping_behavior_updated.csv")
print(df.shape)

drop_cols = ["Customer ID"]
df = df.drop(columns=drop_cols)
df.head(-20)
```

```python
for col in ['Gender', 'Item Purchased', 'Category', 'Location', 'Size', 'Color', 'Season', 'Subscription Status', 'Shipping Type', 'Discount Applied', 'Promo Code Used', 'Payment Method', 'Frequency of Purchases']:
    codes, unique = pd.factorize(df[col])
    df[col] = codes

# payment_method = df.pop("Gender")
# df["Gender"] = payment_method
df.head()
```

```python
df.to_csv("shopping_processed.csv", header=False, index=False)
```
